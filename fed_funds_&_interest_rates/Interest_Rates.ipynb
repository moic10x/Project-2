```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
fedfunds_df = pd.read_csv(
    '../raw_data/DFF.csv'
    )
fedfunds_df
```

```python
fedfunds_df['DATE'] = pd.to_datetime(fedfunds_df['DATE'])
fedfunds_df.set_index('DATE', inplace=True)
fedfunds_df = fedfunds_df.resample('Q').mean()

fedfunds_df.tail()
```

```python
thirty_yr_mtg_df = pd.read_csv('../raw_data/MORTGAGE30US.csv')
```

```python
thirty_yr_mtg_df['DATE'] = pd.to_datetime(thirty_yr_mtg_df['DATE'])
thirty_yr_mtg_df.set_index('DATE', inplace=True)
thirty_yr_mtg_df = thirty_yr_mtg_df.resample('Q').mean()

thirty_yr_mtg_df.tail()
```

```python
rates_df = fedfunds_df.merge(thirty_yr_mtg_df, how='outer', on='DATE')
rates_df.head()
```

```python
rates_df = rates_df.reset_index()
rates_df
```

```python
rates_df['DATE'] = pd.to_datetime(rates_df['DATE']).dt.to_period('M')
rates_df = rates_df.set_index('DATE')
print(rates_df)
```

```python
rates_df.to_csv(Path('../clean_data/clean_interest_rates.csv'))
```
